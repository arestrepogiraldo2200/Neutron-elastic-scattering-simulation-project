```python
def graf(name,sigma,lp,E):
    df=pd.read_csv(name,sep="\t")
    df.theta=df.theta*180/np.pi
    th=df.theta.to_numpy()
    dsigma=df.dsigma.to_numpy()
    plt.plot(th,dsigma)
    #plt.ylim(0,3000)
    plt.xlim(0,180)
    plt.title(r"$\sigma$ = "+str(sigma)+"mb  $l_p$ ="+str(lp)+"  E= "+str(E)+ " MeV")
    plt.grid()
    plt.yscale('log')
    plt.show()
```

```python
# Pa r a m e t e r s :
V1= 56.3
V2= 0.32
V3= 24.0
a= 0.75
r0= 1.17
expmax = 700.0
# [MeV]
# [MeV]
# s u r f a c e t h i c k n e s s p a r a m e t e r [ fm ]
# n e u t r o n r a d i u s [ fm ]
# maximum o f exponent f o r e x p o n e n t i a l f u n c t i o n
# F u n c t i o n f o r t h e c a l c u l a t i o n o f t h e WoodsSaxon p o t e n t i a l :
def v(R, E , A, Z ) :
    V0=V1 -(V2*E) -(V3 * ( 1.0-(( 2.0 * Z)/A) ) )
    r1 = r0 * (A** ( 1.0 / 3.0 ) ) # r a d i u s o f t h e n u c l e u s
# P r e v e n t i o n o f an o v e r f l o w :
    if (( R-r1 )/a)<= expmax:
        V = -V0/(1.0+ np.exp (( R-r1 )/ a ) )
    else :
        V = 0.0
    return V
```
