```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Potenciales para diferentes energías de un mismo isótopo

## $^{40}$Ca

```python
dfCa1 = pd.read_csv("Ca40/Ca40_1MeV_Potencial.txt" ,sep="\t")
dfCa10 = pd.read_csv("Ca40/Ca40_10MeV_Potencial.txt" ,sep="\t")
dfCa20 = pd.read_csv("Ca40/Ca40_20MeV_Potencial.txt" ,sep="\t")
dfCa30 = pd.read_csv("Ca40/Ca40_30MeV_Potencial.txt" ,sep="\t")
dfCa40 = pd.read_csv("Ca40/Ca40_40MeV_Potencial.txt" ,sep="\t")
dfCa50 = pd.read_csv("Ca40/Ca40_50MeV_Potencial.txt" ,sep="\t")

fig = plt.figure(figsize = (8,6))

plt.plot(dfCa1.r, dfCa1['V(r)'], label = '1 MeV')
plt.plot(dfCa10.r, dfCa10['V(r)'], label = '10 MeV')
plt.plot(dfCa20.r, dfCa20['V(r)'], label = '20 MeV')
plt.plot(dfCa30.r, dfCa30['V(r)'], label = '30 MeV')
plt.plot(dfCa40.r, dfCa40['V(r)'], label = '40 MeV')
plt.plot(dfCa50.r, dfCa50['V(r)'], label = '50 MeV')

plt.xlabel('r [fm]', fontsize= 20)
plt.ylabel('V(r) [MeV]', fontsize= 20)
plt.grid(alpha = 0.3)
plt.legend()
plt.text(15,-50,'$^{40}Ca$', fontsize = 30)
plt.savefig('40CaPot.png', dpi = 500)

plt.show()
```
